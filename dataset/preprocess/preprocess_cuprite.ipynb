### Preprocess cuprite dataset

```python
import scipy.io as sio
```

```python
cuprite = sio.loadmat("../source_datasets/cuprite/cuprite_ref.mat")
cuprite_end = sio.loadmat("../source_datasets/cuprite/cuprite_end.mat")

print(cuprite.keys())
print(cuprite_end.keys())
```

```python
cuprite["X"].shape, cuprite_end["M"].shape, cuprite_end["nEnd"].shape, cuprite_end["cood"].shape
```

```python
cuprite["X"] = cuprite["X"].reshape(512, 614, 188).transpose(2, 0, 1)
print(cuprite["X"].shape)
```

```python
cuprite["M"] = cuprite_end["M"]
print(cuprite["M"].shape)
```

```python
cuprite["P"] = cuprite_end["nEnd"].item()
print(cuprite["P"])
```

```python
cuprite["H"] = 512
cuprite["W"] = 614
cuprite["C"] = 188
print(cuprite["H"], cuprite["W"], cuprite["C"])
```

```python
print(cuprite.keys())
```

```python
sio.savemat("../cuprite/cuprite.mat", cuprite)
```
