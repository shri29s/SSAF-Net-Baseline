### Preprocess jasper dataset

```python
import scipy.io as sio
```

```python
jasper = sio.loadmat("../source_datasets/jasper/jasperRidge2_R198.mat")
jasper_end = sio.loadmat("../source_datasets/jasper/jasperRidge2_end4.mat")

print(jasper.keys())
print(jasper_end.keys())
```

```python
import numpy as np

for key in jasper.keys():
    print(key, jasper[key].shape if isinstance(jasper[key], np.ndarray) else jasper[key])

for key in jasper_end.keys():
    print(key, jasper_end[key].shape if isinstance(jasper_end[key], np.ndarray) else jasper_end[key])
```

```python
print(jasper["R"].shape)
```

```python
jasper["Y"] = jasper["R"].reshape(198, 100, 100)
del jasper["R"]
```

```python
print(jasper["Y"].shape)
```

```python
print(jasper["Region"][0, 0][0])
print(jasper["SlectBands"])
```

```python
del jasper["Region"]
del jasper["SlectBands"]
```

```python
jasper.keys()
```

```python
print(jasper["nRow"].item())
print(jasper["nCol"].item())
```

```python
jasper["H"] = jasper["nRow"].item()
jasper["W"] = jasper["nCol"].item()

del jasper["nRow"]
del jasper["nCol"]
del jasper["maxValue"]
del jasper["nBand"]
```

```python
print(jasper.keys())
```

```python
print(jasper_end.keys())
```

```python
del jasper_end["cood"]
```
